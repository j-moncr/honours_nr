```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
G = 6.6743e-11  # gravitational constant
M = 1.9891e30   # mass of the Sun
a = 1.597870049e11  # semi-latus rectum of Mercury's orbit

# Define initial conditions
r0 = 4.60012e10   # distance of Mercury from the Sun at the perihelion
theta0 = np.pi/2  # angle with respect to the z-axis at the perihelion
phi0 = 0          # angle with respect to the x-axis at the perihelion

# Calculate the velocity at the perihelion using the vis-viva equation
v0 = np.sqrt(G * M * (2/r0 - 1/a))

# Define time span and number of time steps for simulation
t_span = (0, 10*365*24*3600)  # 10 years in seconds
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Define initial state vector
y0 = [r0, theta0, phi0, 0, 0, v0]

# Solve equations of motion and get particle's position
params = (M, 0, 0, 1, 0, 0)
x, y, z = solve_EOM(t_span, y0, params, t_eval)

# Plot the trajectory of Mercury
plot_trajectory(x, y, z, params, title="Orbit of Mercury around the Sun")
```
