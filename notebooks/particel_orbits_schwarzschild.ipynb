```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Schwwarzschild effective potential
def V_sc(r, sigma, G=1, M=10, L=10):
    return 1/2 * sigma - sigma * G * M / r + L**2 / (2 * r**2)  - G * M * L**2 / r**3
```

```python
r = np.linspace(2, 10, 1000)
V_sc(r, sigma=1)
```

```python
r = np.linspace(2, 30, 1000)

for L_M in np.linspace(0.5, 5, 5):
    plt.plot(r, V_sc(r, sigma=0, G=1,M=1,L=L_M), label=f"L/M={L_M}")
    # plt.yscale("log")
plt.axvline(3,ymin=0.03,ls="--",c="k", label="3GM")
plt.axvline(2, ymin=0.03, ls="-.", c="k", label="Event horizon")
plt.legend()
plt.xlabel("r")
plt.ylabel("Veff")
plt.title("Effective potetnial for massless particle")
plt.show()
```

```python
for L_M in np.linspace(3, 4, 8):
    plt.plot(r, V_sc(r, sigma=1, G=1, M=1, L=L_M), label=f"L/M={L_M:.2f}")
# L_M = 10
# plt.plot(r, V_sc(r, sigma=1, G=1, M=1, L=L_M), label=f"L/M={L_M:.2f}")
plt.legend()
plt.xlabel("r")
plt.ylabel("Veff")
plt.ylim(0.40, 0.51)
plt.xlim(2, 35)
plt.title("Effective potential for massive particle")
plt.show()
```

# Kerr space-time

```python
# Kerr effective potential for massive particle
def V_kerr(r, a, G=1, M=1, l=1, e=1):
    return -M/r  + (l**2 - a**2 * (e**2 - 1)) / (2*r**2) - M *(l-a*e)**2 / r**3
```

```python
# r = np.linspace(2, 30, 1000)



for L_M in np.linspace(3, 4, 8):
    plt.plot(r, V_kerr(r, 1, G=1, M=1, l=L_M), label=f"L/M={L_M:.2f}")
# L_M = 10
# plt.plot(r, V_sc(r, sigma=1, G=1, M=1, L=L_M), label=f"L/M={L_M:.2f}")
plt.legend()
plt.xlabel("r")
plt.ylabel("Veff")
plt.ylim(-0.15, 0.01)
plt.xlim(2, 35)
plt.title("Effective potential for massive particle")
plt.show()
```

```python
r = np.linspace(2, 35, 100)

fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(16, 8.))

for i, a in enumerate(np.linspace(0, 0.8, 5)):
    for L_M in np.linspace(3, 4, 8):
        axs[0][i].plot(r, V_kerr(r, a, G=1, M=1, l=L_M), label=f"L/M={L_M:.2f}")  
    
    axs[0][i].axvline(2, ls="--", c="k")
    axs[0][i].set_title(f"a={a:.1f}")
    axs[0][i].set_xlabel("r")
    axs[0][i].set_ylabel("Veff")
    axs[0][i].set_ylim(-0.1, 0.15)
    axs[0][i].set_xlim(0.5, 35)

for i, a in enumerate(np.linspace(-0.2, -1.0, 5)):
    for L_M in np.linspace(3, 4, 8):
        axs[1][i].plot(r, V_kerr(r, a, G=1, M=1, l=L_M), label=f"L/M={L_M:.2f}")  
    
    axs[1][i].axvline(2, ls="--", c="k")
    axs[1][i].set_title(f"a={a:.1f}")
    axs[1][i].set_xlabel("r")
    axs[1][i].set_ylabel("Veff")
    axs[1][i].set_ylim(-0.1, 0.0)
    axs[1][i].set_xlim(0.5, 35)

plt.suptitle("Effective potential for Kerr black hole (Massive particles orbiting equator)")
plt.show()
```

```python
r = np.linspace(2, 35, 100)

fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

for i, a in enumerate(np.linspace(-0.2, -1.0, 5)):
    for L_M in np.linspace(3, 4, 8):
        axs[i].plot(r, V_kerr(r, a, G=1, M=1, l=L_M), label=f"L/M={L_M:.2f}")  
    
    axs[i].axvline(2, ls="--", c="k")
    axs[i].set_title(f"a={a:.1f}")
    axs[i].set_xlabel("r")
    axs[i].set_ylabel("Veff")
    axs[i].set_ylim(-0.2, 0.2)
    axs[i].set_xlim(0.5, 35)

plt.suptitle("Effective potential for Kerr black hole with varying a")
plt.show()
```

```python
def W_kl(r, a, G=1, M=1, b=1, sigma=1):
    return 1/(r**2) * (1 - (a/b)**2 - 2*M/r * (1 - sigma * (a/b))**2)
```

```python
r = np.linspace(2, 35, 10000)

fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

for i, a in enumerate(np.linspace(0.0, 1.0, 5)):
    for L_M in np.linspace(3, 4, 8):
        axs[i].plot(r, W_kl(r, a, G=1, M=1, b=L_M, sigma = np.abs(L_M)/L_M), label=f"L/M={L_M:.2f}")

    axs[i].axvline(2, ls="--", c="k")
    axs[i].set_title(f"a={a:.1f}")
    axs[i].set_xlabel("r")
    axs[i].set_ylabel("Veff")
    axs[i].set_ylim(0, 0.1)
    axs[i].set_xlim(0.5, 10)

plt.suptitle("Effective potential for Kerr black hole (massless orbits)")
plt.show()
```
